```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("../raw_data/weather/weather_madrid.csv")
df.head(3)
```

```python
df["time"] = 2
for i in range(len(df['dt_iso'])):
    item=df['dt_iso'][i]
    groups= item.split(" ")
    date = groups[0]
    raw_time = groups[1].split(":")
    time = raw_time[0]
    df["time"][i] = time
    
```

```python

df['day'] = df.apply(lambda x:datetime.strptime(x['dt_iso'], '%Y-%m-%d %H:%M:%S +0000 UTC').date(), axis=1)
df.head(6)
```

```python
df["datetime"] = df["day"].astype(str)+" "+df["time"].astype(str)
df.head(2)
```

```python
cleaned_df = df[["feels_like","weather_main","datetime"]]
cleaned_df.head()
```
